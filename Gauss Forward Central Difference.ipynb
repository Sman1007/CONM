```python
#Gauss forward Central Difference interpolation
#This code will work for both odd and even points

def Div_Diff_Table(x, f, n):
    for j in range(1, n):
        for i in range(n-j):
            f[i][j] = (f[i+1][j-1] - f[i][j-1])
    return f

def Print_Diff_Table(f, n):
    for i in range(n):
        for j in range(n - i):
            print(round(f[i][j], 4), end="\t ");
 
        print("");
    
n = int(input("Enter number of points: "))    
x = list(map(int, input("x: ").split() ))       # input x and y values

h = x[1] - x[0]
f = [[0 for i in range(n)]
        for j in range(n)];

for i in range(n):
    f[i][0] = int(input("Enter f[{}]: ".format(i)))
    
f=Div_Diff_Table(x, f, n);

# displaying divided difference table
print("Divided difference Table:")
Print_Diff_Table(f, n)

k = (n-1)//2
c = 2
a=[]
for i in range(n):
    if c == 0:
        c = 2
        k = k-1
    a.append((f[k][i]))
    c = c - 1
print("Del f terms: ", a)   
'''for i in range(0, n, 2):  will work as same'''

def prod(x, val, i, n):
    p = 1
    m = int((n-1)/2)
    down = 0
    up = 1
    for j in range(i):
        if j >= 1 and j%2 == 0:
            p = p*(val - x[m - up])*(1/h)*(1/(j+1))
            up = up + 1
        else:
            p = p*(val - x[m+down])*(1/h)*(1/(j+1))
            down += 1
    return p

def Cen_Div_Diff(x, f, value, n):
    m = int((n-1)/2)
    s = a[0]
    for i in range(1,n):
        s = s+ a[i]*prod(x, value, i, n)
    return s

k = float(input("Enter interpolating point: "))

y = Cen_Div_Diff(x, f, k, n)
print("Approximated value of f({}) is {}".format(k, y))
```
