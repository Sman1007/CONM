```python
import math 

def y(x):
    f = (math.sqrt(1+4*(x**2)))*(math.exp(-(x**2)))
    return f

# Inputs
n = int(input("Enter number of intervals: "))
x0, xn = map(int, input("Enter x0 and xn: ").split())
 
h = (xn-x0)/n                         # step size

def Trapizoidal(y, a, b, n):          # Trapizoidal 
    A = y(a) + y(b)
    z = h
    for i in range(1,n):
        A = A + 2*y(a+z)
        z=z+h
    A = A*(h/2)
    return A

def SimpsonsOneThird(y, a, b, n):     # Simpson's OneThird
    A1 = y(a) + y(b)
    z1 = h
    if n%2 == 0:
        for i in range(1,n):
            if i%2 == 1:
                A1 = A1 + 4*y(a+z1)
            else:
                A1 = A1 + 2*y(a+z1)
            z1=z1+h
        A1 = A1*(h/3)
        return A1
    else:
        print("n is not even.")
        
        
def SimpsonsThreeEight(y, a, b, n):  # Simpson's Three by Eight
    A = y(a) + y(b)
    z = h
    if n%3 == 0:
        for i in range(1,n):
            if i%3 != 0 :
                A = A + 3*y(a+z) 
            else:
                A = A + 2*y(a+z)
            z=z+h

        A = A*(h*(3/8))
        return A
    else:
        print("n is not multiple of 3.")

        
# Defining phi function for Gaussian Quadrature

def phi(t, a, b):
    f = ( (b-a)/2 )*( y(((b-a)/2)*t + (b+a)/2) )
    return f

def GaussQuadrature(a, b):                         # Gaussian Quadrature
    A = phi(1/(3**(1/2)), a, b) + phi(-1/(3**(1/2)), a, b)
    return A
    
T = Trapizoidal(y, x0, xn, n)
S_13 = SimpsonsOneThird(y, x0, xn, n)
S_38 = SimpsonsThreeEight(y, x0, xn, n)
G_Q = GaussQuadrature(x0, xn)

print("\nTrapizoidal: ", T)
print("SimpsonsOneThird: ", S_13)
print("SimpsonsThreeEight: ", S_38)
print("GaussQuadrature: ", G_Q)

from scipy.integrate import quad                   # Importing integration and calculate integration
k = quad(y, x0, xn)
act = k[0]

print("\nActual integration by python library: ", act)

print("\nTrapizoidal error: ", abs(T - act))       # Printing Errors for each methods
print("SimpsonsOneThird error: ", abs(S_13 - act))
print("SimpsonsThreeEight error: ", abs(S_38 - act))
print("GaussQuadrature error: ", abs(G_Q - act))
```
